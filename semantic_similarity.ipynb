```python
from sentence_transformers import SentenceTransformer, util
import numpy as np
```

```python
model = SentenceTransformer('stsb-roberta-large')
```

```python
s1 = "Humpty Dumpty sat on a wall"
s2 = "Humpty Dumpty had a great fall"
s3 = "Robert DeNiro sat on a wall"
```

```python
e1 = model.encode(s1, convert_to_tensor=True)
e2 = model.encode(s2, convert_to_tensor=True)
e3 = model.encode(s3, convert_to_tensor=True)
```

```python
util.pytorch_cos_sim(e1, e2)
```

```python
util.pytorch_cos_sim(e2, e3)
```

```python
util.pytorch_cos_sim(e1, e3)
```

```python
lines = lines[8:-1]
```

```python
all_words_embeddings = model.encode(lines, convert_to_tensor=True)
```

```python
word = "to the right"
selected_embedding = model.encode(word, convert_to_tensor=True)

cos_scores = util.pytorch_cos_sim(selected_embedding, all_words_embeddings)[0]

num_results = 10
top_results = np.argpartition(-cos_scores, range(num_results))[0:num_results]
print("The next ajective shoudl involve the words:")
for idx in top_results[0:num_results]:
    print(lines[idx], "(Score: %.4f)" % (cos_scores[idx]))
```
